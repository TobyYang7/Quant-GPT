```python
import pandas as pd
from tqdm import tqdm
from transformers import AutoTokenizer

model_name = "/Users/yuzheyang/.cache/modelscope/hub/qwen/Qwen-7B-Chat"
tokenizer = AutoTokenizer.from_pretrained(model_name,  trust_remote_code=True)


def calculate_token_length(text):
    model_name = "/Users/yuzheyang/.cache/modelscope/hub/qwen/Qwen-7B-Chat"
    tokenizer = AutoTokenizer.from_pretrained(model_name,  trust_remote_code=True)
    tokens = tokenizer.encode(text, truncation=True)
    return len(tokens)


stock_list = ["600031.SH", "600036.SH", "600050.SH", "600104.SH", "600346.SH", "600570.SH", "600887.SH", "601390.SH", "603160.SH", "601668.SH"]

for stock in tqdm(stock_list):
    df = pd.read_parquet(f'/Users/yuzheyang/dev/Quant-GPT/data/announcement/{stock}.parquet')
    df['token_count'] = df['content'].apply(calculate_token_length)
    df.to_parquet(f'/Users/yuzheyang/dev/Quant-GPT/data/token/{stock}_token.parquet')
```
