```python
import pandas as pd
from tqdm import tqdm

data = pd.read_excel("A_news_train_last1000.xlsx")
data
```

```python
import json
from tqdm import tqdm

results = {}  # 继续使用字典存储
for i, row in tqdm(enumerate(data.iterrows())):
    label_mapping = {0: "极度负面", 1: "负面", 2: "中性", 3: "正面", 4: "极度正面"}
    label = label_mapping[row[1]['label']]
    query = row[1]['content']
    prompt = f"这是一篇来自新浪财经的新闻稿，请判断该新闻的市场情绪属于以下哪一类别 (极度负面/负面/中性/正面/极度正面)"

    results[i] = {
        "label": label,
        "query": query,
        "prompt": prompt
    }

# 字典转列表
results_list = list(results.values())

# 切分数据
train_size = int(0.85 * len(results_list))
train_data = results_list[:train_size]
test_data = results_list[train_size:]

# 写入JSON文件
with open('sentiment/train.json', 'w') as f:
    json.dump(train_data, f, indent=4, ensure_ascii=False)

with open('sentiment/test.json', 'w') as f:
    json.dump(test_data, f, indent=4, ensure_ascii=False)
```
